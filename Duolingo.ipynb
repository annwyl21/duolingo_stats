```python
import datetime 
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

#Dataframe storing information about my language studies
df = pd.DataFrame({
  'Language': ['Dutch', 'Welsh', 'Scots Gaelic', 'German', 'Spanish', 'French', 'Swedish'],
  #'Irish', 'Russian', 'Greek', 'Italian', 'Polish', 'Swahili', 'Mandarin', 'Japanese'
  'Duolingo Checkpoint Passed': [2, 1, 1, 2, 2, 3, 0],
  'Crowns': [203, 102, 81, 175, 288, 226, 4],
  'Words': [1056, 465, 398, 444, 1518, 1867, 25],
  })
#extra lines added to dataframe
df['Duolingo XP'] = [18588, 10730, 11819, 14163, 41757, 22829, 641]
df['Days Active'] = [189, 131, 126, 130, 302, 271, 4]
df['Number of Lessons Completed'] = [1206, 699, 772, 782, 1722, 1355, 21]

#import csv file
df_leaderboard = pd.read_csv('duolingo_leaderboards.csv')
#print(df_leaderboard.head())
```

This notebook contains my progress data downloaded from Duolingo and documenting my use of Duolingo to learn several languages over the past few years.

```python
sns.set_style("darkgrid")
sns.set_palette("muted")#deep, muted, pastel, bright, dark, and colorblind
sns.set_context('notebook', font_scale=0.7)
```

I am using Matplotlib and Seaborn to create graphs to investigate my use of Duolingo, and Pandas to create the dataframe that contains some of the data I obtained from my profile rather than in the downloaded csv file.

```python
#Bar Chart Languages Learned and Level Achieved
ax1=plt.figure(figsize=(5,4))
ax1=plt.subplot()
ax1=sns.barplot(data=df, x='Language', y='Duolingo Checkpoint Passed')
ax1=plt.title('Current Level in Languages Learned')
# ax1.set_yticks((1, 2, 3))
# ax1.set_yticklabels(('Level 1', 'Level 2', 'Level 3'))
```

I have been learning 7 languages over the past few years in addition to Python and Javascript. This bar chart shows the levels I have achieved in each language, while caring for my family and running the household.

```python
#Pie Chart Crowns Earned in each Language
ax2=plt.figure(figsize=(6,4))
ax2=plt.subplot()
ax2=plt.pie(df['Crowns'], autopct='%d%%')
ax2=plt.axis('equal')
ax2=plt.title("Duolingo Crowns Earned in each Language")
ax2=plt.legend(df['Language'], loc='upper left')
```

This pie chart shows the percentage of crowns I have achieved in each language. It looks as though I have worked hardest in Spanish.
Welsh and Scots Gaelic are super interesting languages but they are longer-term goals.

```python
#Pie Chart XP
ax3=plt.figure(figsize=(6,4))
ax3=plt.subplot()
ax3=plt.pie(df['Duolingo XP'], autopct='%d%%')
ax3=plt.axis('equal')
ax3=plt.title('Duolingo XP earned in each Language')
ax3=plt.legend(df['Language'], loc='upper left')
```

But this pie chart presents a better picture of my recent learning. I have focused alot on Spanish this year.

```python
# #Words learned by level of vocabulary in each language
ax4=plt.figure(figsize=(6,5))
ax4=plt.subplot()
ax4=sns.barplot(data=df, x='Language', y='Words', hue='Duolingo Checkpoint Passed')
ax4=plt.title('Complexity of Vocabulary')
```

This bar graph shows how many words I have learnt in relation to where I am in the Duolingo study levels.

```python
#Distribution of League Scores
#KDE distribution
ax5=sns.kdeplot(df_leaderboard['score'], shade=True)
```

I seem to either make 1,000 points a week or really go for it and make 4,000 point per week. I think that does acurately reflect my 'ticking along' weeks and my 'going for it' weeks.

```python
# box plots, showing outliers
ax6=sns.boxplot(data=df_leaderboard, x='score')
```

My average weekly points score is just under 1500 points.

```python
# violin plot
ax7=sns.violinplot(data=df_leaderboard, x='score')
```

There is a clear 'going for it' bump in the violin plot for those weeks when I try harder or we are in Lockdown.

```python
#Scores in each League
ax8=sns.barplot(data=df_leaderboard, x='tier', y='score')
ax8=plt.title('How does each leaderboard influence my weekly score?')
```

```python
#Scatterplot to show days active against the number of lessons completed
ax9=sns.scatterplot(x=df['Days Active'], y=df['Number of Lessons Completed'], hue=df['Language'])
ax9=plt.title('How does length of time learning affect the number of lessons I complete?')

plt.show()
```
